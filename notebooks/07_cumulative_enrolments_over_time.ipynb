```python
import pandas as pd
import matplotlib.pyplot as plt

enrolment_df=pd.read_parquet("../sql/data/data_cache/enrolment_clean.parquet")

enrolment_df["date"]=pd.to_datetime(enrolment_df["date"])
daily=enrolment_df.groupby("date").size().reset_index(name="daily_enrolments")
daily["cumulative_enrolments"]=daily["daily_enrolments"].cumsum()

plt.figure(figsize=(10,4))
plt.plot(daily["date"],daily["cumulative_enrolments"])
plt.xlabel("date")
plt.ylabel("cumulative enrolments")
plt.title("cumulative aadhaar enrolments over time")
plt.tight_layout()
plt.show()
```
