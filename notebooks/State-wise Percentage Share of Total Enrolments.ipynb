```python
import pandas as pd
import matplotlib.pyplot as plt

enrolment_df=pd.read_parquet("../sql/data/data_cache/enrolment_clean.parquet")

state_enrolments=enrolment_df.groupby("state").size()
percentage=(state_enrolments/state_enrolments.sum())*100

percentage.sort_values(ascending=False).plot(kind="bar")

plt.xlabel("state")
plt.ylabel("percentage share (%)")
plt.title("state-wise percentage share of total enrolments")
plt.tight_layout()
plt.show()
```
