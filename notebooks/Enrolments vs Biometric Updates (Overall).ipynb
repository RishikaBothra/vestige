```python
import pandas as pd
import matplotlib.pyplot as plt

enrolment_df=pd.read_parquet("../sql/data/data_cache/enrolment_clean.parquet")
biometric_df=pd.read_parquet("../sql/data/data_cache/biometric_clean.parquet")

counts={
    "enrolments":len(enrolment_df),
    "biometric updates":len(biometric_df)
}

plt.figure()
plt.bar(counts.keys(),counts.values())
plt.ylabel("count")
plt.title("enrolments vs biometric updates (overall)")
plt.tight_layout()
plt.show()
```
