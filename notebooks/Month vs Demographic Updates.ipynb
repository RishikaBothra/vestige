```python
import pandas as pd
import matplotlib.pyplot as plt

demographic_df=pd.read_parquet("../sql/data/data_cache/demographic_clean.parquet")

demographic_df["month"]=pd.to_datetime(demographic_df["date"]).dt.to_period("M").astype(str)

month_demographic=(
    demographic_df
    .groupby("month")
    .size()
    .sort_index()
)

plt.figure()
plt.plot(month_demographic.index,month_demographic.values)
plt.xticks(rotation=45)
plt.yscale("log")
plt.xlabel("month")
plt.ylabel("demographic updates")
plt.title("month vs demographic updates")
plt.tight_layout()
plt.show()
```
