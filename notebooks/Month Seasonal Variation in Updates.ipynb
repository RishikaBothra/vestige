```python
import pandas as pd
import matplotlib.pyplot as plt

demographic_df=pd.read_parquet("../sql/data/data_cache/demographic_clean.parquet")
biometric_df=pd.read_parquet("../sql/data/data_cache/biometric_clean.parquet")

updates=pd.concat([demographic_df,biometric_df])
updates["month_num"]=pd.to_datetime(updates["date"]).dt.month

seasonal=updates.groupby("month_num").size()

plt.figure()
seasonal.plot(kind="bar")
plt.xlabel("month (1=jan, 12=dec)")
plt.ylabel("updates")
plt.title("seasonal variation in updates")
plt.tight_layout()
plt.show()
```
