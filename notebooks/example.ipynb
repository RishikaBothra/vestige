```python
import pandas as pd

raw=pd.read_excel("../sql/data/AadhaarSaturationReport.xlsx",header=None)

# extract proper table
df=raw.iloc[5:41].copy()
df.columns=[
    "sno",
    "state",
    "population_2025",
    "aadhaar_assigned",
    "penetration_ratio"
]

# clean numbers
df["population_2025"]=df["population_2025"].astype(str).str.replace(",","").astype(int)
df["aadhaar_assigned"]=df["aadhaar_assigned"].astype(str).str.replace(",","").astype(int)
df["penetration_percent"]=df["penetration_ratio"].astype(float)*100

# drop serial + ratio column
df=df.drop(columns=["sno","penetration_ratio"])

print("shape:",df.shape)
print(df.head())
print("\nmin / max penetration:")
print(df["penetration_percent"].min(),df["penetration_percent"].max())
```
