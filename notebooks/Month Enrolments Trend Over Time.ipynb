```python
import pandas as pd
import matplotlib.pyplot as plt

enrolment_df=pd.read_parquet("../sql/data/data_cache/enrolment_clean.parquet")

enrolment_df["month"]=pd.to_datetime(enrolment_df["date"]).dt.to_period("M").astype(str)

trend=enrolment_df.groupby("month").size().sort_index().cumsum()

plt.figure()
plt.plot(trend.index,trend.values)
plt.yscale("log")
plt.xticks(rotation=45)
plt.xlabel("month")
plt.ylabel("cumulative enrolments")
plt.title("enrolments trend over time")
plt.tight_layout()
plt.show()
```
