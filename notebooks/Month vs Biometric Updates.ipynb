```python
import pandas as pd
import matplotlib.pyplot as plt

biometric_df=pd.read_parquet("../sql/data/data_cache/biometric_clean.parquet")

biometric_df["month"]=pd.to_datetime(biometric_df["date"]).dt.to_period("M").astype(str)

month_biometric=(
    biometric_df
    .groupby("month")
    .size()
    .sort_index()
)

plt.figure()
plt.plot(month_biometric.index,month_biometric.values)
plt.xticks(rotation=45)
plt.xlabel("month")
plt.ylabel("biometric updates")
plt.title("month vs biometric updates")
plt.tight_layout()
plt.show()
```
