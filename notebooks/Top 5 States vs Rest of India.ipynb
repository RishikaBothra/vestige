```python
import pandas as pd
import matplotlib.pyplot as plt

enrolment_df=pd.read_parquet("../sql/data/data_cache/enrolment_clean.parquet")

counts=enrolment_df.groupby("state").size().sort_values(ascending=False)

top5=counts.head(5).sum()
rest=counts.iloc[5:].sum()

plt.figure()
plt.pie([top5,rest],labels=["top 5 states","rest of india"],autopct="%1.1f%%")
plt.title("top 5 states vs rest of india (enrolments)")
plt.show()
```
