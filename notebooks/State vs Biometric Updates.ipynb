```python
import pandas as pd
import matplotlib.pyplot as plt

# load data
biometric_df=pd.read_parquet("../sql/data/data_cache/biometric_clean.parquet")

# aggregate state-wise biometric updates
state_biometric=(
    biometric_df
    .groupby("state")
    .size()
    .sort_values(ascending=False)
)
# plot
plt.figure()
state_biometric.plot(kind="bar")
plt.xlabel("state")
plt.yscale("log") 
plt.ylabel("biometric updates")
plt.title("state vs biometric updates")
plt.tight_layout()
plt.show()
```
