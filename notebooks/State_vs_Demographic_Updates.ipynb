```python
import pandas as pd
import matplotlib.pyplot as plt

# load data
demographic_df=pd.read_parquet("../sql/data/data_cache/demographic_clean.parquet")

# state-wise demographic update count
state_demographic=(
    demographic_df
    .groupby("state")
    .size()
    .sort_values(ascending=False)
)

# plot
plt.figure()
state_demographic.plot(kind="bar")
plt.xlabel("state")
plt.yscale("log") 
plt.ylabel("demographic updates")
plt.title("state vs demographic updates")
plt.tight_layout()
plt.show()
```
