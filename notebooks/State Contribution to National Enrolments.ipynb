```python
import pandas as pd
import matplotlib.pyplot as plt

enrolment_df=pd.read_parquet("../sql/data/data_cache/enrolment_clean.parquet")

counts=enrolment_df.groupby("state").size()
share=(counts/counts.sum())*100

share.sort_values(ascending=False).plot(kind="bar")
plt.ylabel("percentage share")
plt.yscale("log")
plt.title("state contribution to national enrolments")
plt.tight_layout()
plt.show()
```
