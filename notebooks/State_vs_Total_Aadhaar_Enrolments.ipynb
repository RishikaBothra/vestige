```python
import pandas as pd
import matplotlib.pyplot as plt


enrolment_df=pd.read_parquet("../sql/data/data_cache/enrolment_clean.parquet")


state_enrolments=(
    enrolment_df
    .groupby("state")
    .size()
    .sort_values(ascending=False)
)

# plot
plt.figure()
state_enrolments.plot(kind="bar")
plt.xlabel("state")
plt.yscale("log") 
plt.ylabel("total enrolments")
plt.title("state vs total aadhaar enrolments")
plt.tight_layout()
plt.show()
```
