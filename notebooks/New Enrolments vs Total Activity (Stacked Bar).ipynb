```python
import pandas as pd
import matplotlib.pyplot as plt

enrolment_df=pd.read_parquet("../sql/data/data_cache/enrolment_clean.parquet")
demographic_df=pd.read_parquet("../sql/data/data_cache/demographic_clean.parquet")
biometric_df=pd.read_parquet("../sql/data/data_cache/biometric_clean.parquet")

enrolments=len(enrolment_df)
updates=len(demographic_df)+len(biometric_df)

plt.figure()
plt.bar(["total activity"],[enrolments],label="enrolments")
plt.bar(["total activity"],[updates],bottom=[enrolments],label="updates")
plt.ylabel("count")
plt.title("new enrolments vs total activity")
plt.legend()
plt.tight_layout()
plt.show()
```
