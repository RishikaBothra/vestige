```python
import pandas as pd
import matplotlib.pyplot as plt

demographic_df=pd.read_parquet("../sql/data/data_cache/demographic_clean.parquet")
biometric_df=pd.read_parquet("../sql/data/data_cache/biometric_clean.parquet")

state_updates=pd.concat([demographic_df,biometric_df]).groupby("state").size()
percentage=(state_updates/state_updates.sum())*100

percentage.sort_values(ascending=False).plot(kind="bar")

plt.xlabel("state")
plt.ylabel("percentage share (%)")
plt.title("state-wise percentage share of total updates")
plt.tight_layout()
plt.show()
```
