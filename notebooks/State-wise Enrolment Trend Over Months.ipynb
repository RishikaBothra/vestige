```python
import pandas as pd
import matplotlib.pyplot as plt

enrolment_df=pd.read_parquet("../sql/data/data_cache/enrolment_clean.parquet")
enrolment_df["month"]=pd.to_datetime(enrolment_df["date"]).dt.to_period("M")

trend=enrolment_df.groupby(["state","month"]).size().unstack(fill_value=0)

trend.T.plot(legend=False)
plt.title("state-wise enrolment trend")
plt.tight_layout()
plt.show()

```
