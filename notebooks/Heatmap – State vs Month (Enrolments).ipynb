```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

enrolment_df=pd.read_parquet("../sql/data/data_cache/enrolment_clean.parquet")
enrolment_df["month"]=pd.to_datetime(enrolment_df["date"]).dt.to_period("M")

pivot=enrolment_df.groupby(["state","month"]).size().unstack(fill_value=0)

plt.figure()
sns.heatmap(pivot)
plt.title("heatmap: enrolments by state and month")
plt.show()
```
