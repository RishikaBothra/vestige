```python
import pandas as pd
import matplotlib.pyplot as plt

enrolment_df=pd.read_parquet("../sql/data/data_cache/enrolment_clean.parquet")

cum=enrolment_df.groupby("state").size().sort_values(ascending=False).cumsum()

plt.figure()
cum.plot()
plt.ylabel("cumulative enrolments")
plt.title("cumulative enrolments by state")
plt.tight_layout()
plt.show()
```
