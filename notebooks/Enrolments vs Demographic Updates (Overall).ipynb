```python
import pandas as pd
import matplotlib.pyplot as plt

enrolment_df=pd.read_parquet("../sql/data/data_cache/enrolment_clean.parquet")
demographic_df=pd.read_parquet("../sql/data/data_cache/demographic_clean.parquet")

counts={
    "enrolments":len(enrolment_df),
    "demographic updates":len(demographic_df)
}

plt.figure()
plt.bar(counts.keys(),counts.values())
plt.ylabel("count")
plt.title("enrolments vs demographic updates (overall)")
plt.tight_layout()
plt.show()
```
