```python
import pandas as pd
import matplotlib.pyplot as plt

enrolment_df=pd.read_parquet("../sql/data/data_cache/enrolment_clean.parquet")
enrolment_df["date"]=pd.to_datetime(enrolment_df["date"])
enrolment_df["month"]=enrolment_df["date"].dt.month

seasonal=enrolment_df.groupby("month").size().reset_index(name="enrolments")

plt.figure(figsize=(8,4))
plt.bar(seasonal["month"],seasonal["enrolments"])
plt.xlabel("month")
plt.ylabel("enrolments")
plt.title("seasonal enrolment pattern (month of year)")
plt.tight_layout()
plt.show()
```
