```python
import pandas as pd
import matplotlib.pyplot as plt

demographic_df=pd.read_parquet("../sql/data/data_cache/demographic_clean.parquet")
biometric_df=pd.read_parquet("../sql/data/data_cache/biometric_clean.parquet")

state_updates=(
    pd.concat([demographic_df,biometric_df])
    .groupby("state")
    .size()
    .sort_values(ascending=False)
)

plt.figure(figsize=(14, 6))
ax = state_updates.plot(kind="bar", color="#4C78A8")
ax.set_yscale("log")
ax.set_xlabel("state")
ax.set_ylabel("total updates (log scale)")
ax.set_title("state vs total updates")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("../outputs/figures/state_total_updates.png", dpi=300)
plt.show()
```
