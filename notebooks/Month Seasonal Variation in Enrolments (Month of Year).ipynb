```python
import pandas as pd
import matplotlib.pyplot as plt

enrolment_df=pd.read_parquet("../sql/data/data_cache/enrolment_clean.parquet")

enrolment_df["month_num"]=pd.to_datetime(enrolment_df["date"]).dt.month

seasonal=enrolment_df.groupby("month_num").size()

plt.figure()
seasonal.plot(kind="bar")
plt.xlabel("month (1=jan, 12=dec)")
plt.ylabel("enrolments")
plt.title("seasonal variation in enrolments")
plt.tight_layout()
plt.show()
```
