```python
import pandas as pd
import matplotlib.pyplot as plt

demo=pd.read_parquet("../sql/data/data_cache/demographic_clean.parquet")
bio=pd.read_parquet("../sql/data/data_cache/biometric_clean.parquet")

updates=pd.concat([demo,bio])
updates["date"]=pd.to_datetime(updates["date"])
updates["month"]=updates["date"].dt.month

seasonal=updates.groupby("month").size().reset_index(name="updates")

plt.figure(figsize=(8,4))
plt.bar(seasonal["month"],seasonal["updates"])
plt.xlabel("month")
plt.ylabel("updates")
plt.title("seasonal update pattern (month of year)")
plt.tight_layout()
plt.show()
```
